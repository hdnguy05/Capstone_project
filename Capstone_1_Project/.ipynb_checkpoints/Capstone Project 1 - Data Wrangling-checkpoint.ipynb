# Read data and Merge data.

```python
import pandas as pd
import numpy as np
from glob import glob
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

Loading all the file of data

```python
file_item = pd.read_csv('competitive-data-science-predict-future-sales/items.csv')
file_categories = pd.read_csv('competitive-data-science-predict-future-sales/item_categories.csv')
file_shop = pd.read_csv('competitive-data-science-predict-future-sales/shops.csv')
file_sale = pd.read_csv('competitive-data-science-predict-future-sales/sales_train.csv', parse_dates = ['date'])
file_test = pd.read_csv('competitive-data-science-predict-future-sales/test.csv')
```

Before we merge data, we looking back the question we need to answer in this data set
  1. Predict totals sales for each item (was provided in test file) next month?
  2. Predit the top 5 category will sold in next month?
  3. Predict the range of price is customer likely to buy?


# Processing merge data

```python
#add month and year column to file_sale
file_sale['month'] = file_sale['date'].dt.month
file_sale['year']  = file_sale['date'].dt.year
```

```python
#Merge item and sale together.
file1 = pd.merge(file_item,file_categories)
file1.head()
```

```python
#Merge category and sale together.
file2 = pd.merge(file_shop,file_sale)
file2.head()
```

```python
#Final file--- merging all 4 file together
master_file = pd.merge(file1,file2)
master_file.head()
```

# Inspecting data

Checking if data have any missing.

```python
missing = list(master_file.isnull().sum().sort_values(ascending = True).items())
missing
```

```python
master_file.describe()
```

Looking at item_cnt_day, and item_price is 2 columns we need for prediction. 
You can note that MIN of this 2 column is negative number. Then we need to remove these data before going next step

```python
#We remove the negative number of these 2 item
master_file = master_file[master_file.item_cnt_day > 0]
master_file = master_file[master_file.item_price > 0]                
master_file.describe()
```

```python

sns.boxplot(master_file['item_cnt_day'])
```

The plot show that we have some outliner, so we need to remove these too

```python
master_file = master_file[master_file.item_cnt_day < 200]
sns.boxplot(master_file['item_cnt_day'])
```

# Preparing data for prediction

QUESTION 1: data prepare

For predict on question 1, we only need these column:['item_id', 'shop_id','date_block_num','item_cnt_day']

```python
question1_data = master_file[['item_id','shop_id','date_block_num','item_cnt_day']]
question1_data = question1_data.groupby(['date_block_num','shop_id','item_id']).agg({'item_cnt_day':'sum'})
question1_data.head(15)
```

Question 2 data:
    

```python

question2_data = master_file[['item_id','shop_id','item_category_id','date_block_num','item_cnt_day','year','month']]
question2_data = question2_data.groupby(['year','month','shop_id','item_category_id']).agg({'item_cnt_day':'sum'})
question2_data.rename(columns={"item_cnt_day":"Total_month_sale"}, errors="raise")
question2_data.head(15)
```

Question 3: Data

```python
question3_data = master_file[['item_id','shop_id','date_block_num','item_cnt_day','item_price']]
question3_data = question3_data.groupby(['date_block_num','shop_id','item_id','item_price']).agg({'item_cnt_day':'sum'})
question3_data.head(15)
```
